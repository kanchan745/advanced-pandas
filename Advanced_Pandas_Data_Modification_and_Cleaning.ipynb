#Modification of data in pandas adding column , update and remove


```python
import pandas as pd
data = {
    "Name": ['Ram','Shyam','Ghanshyam','Rajat','Raj','Kanchan','Kavya'],
    "Age" :[21,34,24,36,25,21,24],
    "Salary":[50000,62000,85000,500000,85000,850000,6200000],
    "Performance_Score":[85,90,70,50,96,75,85]
}
df = pd.DataFrame(data)
print(df)
```

```python
#for assigning newdata in datasets
#square brackets df["Column_Name"] = some_Data(list , series,calculation)
#passing equation

df["Bonus"] = df["Salary"] * 0.1
print(df)

```

```python
#second method : using insert[] method: we can insert a new column at a particular position
df.insert(0,"Employee ID",[10,20,30,40,50,60,70])
print(df)
```

```python
#Updating values
#.loc [] it is a method in pandas  we can access a specific cell and can modify
#we can access cell , column ,set of rows
#SYNTAX df.loc[row_index,"column name"]= new value
#row_index - which have to modifiy


#updating Ram value
df.loc[0,'Salary'] = 55000
print(df)
```

```python
#Updating all values of Salary

#Increasing salary by 5%
df["Salary"] = df["Salary"] * 1.05
print(df)
#
```

```python
#REMOVING COLUMNS : KEEPS YOUR DATA CLEAN AND FOCUSED
# USE DROP METHOD  : WE CAN DELETE ANY COLUMN AND ROY USING DROP METHOD IN ANY DATA FRAME
#SYNTAX : df.drop(columns = ["ColumnName"], inplace = True)
#column name which we want to remove or inplace means itll not modify the data directly (no new data frame ) it ll modify in that data

df.drop(columns = ["Bonus"], inplace = True)
print(df)

```

**for removing multiple columns**


```python
df.drop(columns = ["Performance_Score", "Age"] , inplace = True)
print(df )
```

inplace = True (to modify the originaldata )

Adding Columns Two ways  : directly using assignment , 2.) using insert method

To update values : .loc[] method  and for bulk updation we can combine condition

To remove column : we can use drop method , always place inplace = True
or if multiple columns to remove so (we use , )and column name

dropna() : we can remove or get a value of dataset or where data is mising  

```python
import pandas as pd
Data = {
    "Name": ['Ram','Shyam','Ghanshyam','Rajat','Raj','Kanchan','Kavya'],
    "Age" :[21,None,24,36,25,21,24],
    "Salary":[50000,62000,85000,500000,None,850000,6200000],
    "Performance_Score":[85,90,70,None,96,75,85]
}
df = pd.DataFrame(Data)

df.dropna(axis = 0, inplace = True)
print(df)

# 0 for rows and 1 for columns
```

#Interpollation : follows existing patterns for the fillup of missing values
it finds estimates in missing values in a numerical column
#methods-> linear,polynomial
why use
1) preserve data integrity  = missing values replace with realistic no ,(predict real no.)
2)smooth trends
3)avoid data loss
#interpolate()
# is a method with the help of we can fill missing values in any column and in any dataframe
it finds estimate on the basis of other points which are close

```python
import pandas as pd
data = {
    "Time" : [1,2,3,4,5],
    "Value" : [10,15,None,20,25]
}
df = pd.DataFrame(data)
print('Before Interpolation')
print(df)
df['Value'].interpolate(method = 'linear', inplace = True)
print('After Interpolation')
print(df)
```

when to use
1) when we re working with time series dat
2)numeric data with trends
3) avoid dropping rows
